```python
v1 = sympy.Symbol('v1')
v2 = sympy.Symbol('v2')

def linear_function(v):
    if v >= 0.5:
        return 1
    elif v > -1.5:
        return v
    else:
        return 0

def activation_function(v):
    if v > 0:
        return 1
    elif v == 0:
        return 0
    else:
        return -1
```
